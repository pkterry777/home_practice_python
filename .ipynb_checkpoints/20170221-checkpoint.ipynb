```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import os
import matplotlib.pyplot as plt

#設定字體為標楷體'DFKai-sb':
from pylab import *
 
mpl.rcParams['font.sans-serif'] = ['SimHei']
```

```python
file_name = input('檔案名稱：')
file = open('H:\\實驗室資料\\DATA\\日期\\20170221各區103104年發生率\\%s.csv'%file_name,encoding='big5')
data_list = []
for i in file.readlines():
    data = i.strip()
    data = data.split(',')
    data_list.append(data)
file.close()

file_df = pd.DataFrame(data_list[1:],columns=data_list[0])

condiction = input('是否列印DATAFRAME(y/n)：')
if condiction == 'y':
    print('csv資料：')
    print(file_df)
    print('----------------')
```

```python
dist = list(file_df.ix[:,0])
i103 = list(map(float,list(file_df.ix[:,6])))
i104 = list(map(float,list(file_df.ix[:,7])))
print(dist)
```

```python

fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)
ax1.bar(range(len(dist)),i103,align='center')
ax1.xaxis.set_ticks_position('bottom')
ax1.yaxis.set_ticks_position('left')
plt.xticks(list(range(len(dist))),dist, rotation=90,fontsize=14)
plt.xlabel('里',fontsize=20)
plt.ylabel('發生率',fontsize=20)
plt.title('苓雅區103年登革熱發生率',fontsize=20)
plt.grid(linestyle='--', which='major',color='gray')
plt.show()
```

```python

fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)
ax1.bar(range(len(dist)),i104,align='center')
ax1.xaxis.set_ticks_position('bottom')
ax1.yaxis.set_ticks_position('left')
plt.xticks(list(range(len(dist))),dist, rotation=90,fontsize=14)
plt.xlabel('里',fontsize=20)
plt.ylabel('發生率',fontsize=20)
plt.title('苓雅區104年登革熱發生率',fontsize=20)
plt.grid(linestyle='--', which='major',color='gray')
plt.show()
```

```python
dist_total = []
i103_total = []
i104_total = []
```

```python
dist_total.extend(dist)
i103_total.extend(i103)
i104_total.extend(i104)
```

```python
dist_total.extend([""])
i103_total.extend([0])
i104_total.extend([0])
```

```python
fig = plt.figure(figsize=(25,10))
ax1 = fig.add_subplot(1,1,1)
ax1.bar(range(len(dist_total)),i103_total,align='center')
ax1.xaxis.set_ticks_position('bottom')
ax1.yaxis.set_ticks_position('left')
plt.xticks(list(range(len(dist_total))),dist_total, rotation=90,fontsize=20)
plt.yticks(fontsize=20)
plt.xlabel('三民區          苓雅區           前鎮區',fontsize=30)
plt.ylabel('發生率',fontsize=30)
plt.title('103年登革熱發生率',fontsize=30)
plt.grid(linestyle='--', which='major',color='gray')
plt.show()
```
