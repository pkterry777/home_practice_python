```python
import pandas as pd
import statsmodels.api as sm
import pylab as pl
import numpy as np
```

```python
f1 = open('C:\\Users\\jean\\Desktop\\glm分析\\cluster分析\\ClusterLogistic分析_EML+EMvsE.csv','r')
data_list = []
for i in f1.readlines():
    data = i.strip()
    data = data.split(',')
    data_list.append(data)
f1.close()
GData = pd.DataFrame(data_list[1:],columns=data_list[0])
GData['intercept'] = 1.0
GData.head()
```

```python
X = ['人口數','土地面積','人口密度','性別比','扶老比','扶幼比','扶養比','老化指數','老年人口比例','103年外境Case數']

for i in X:
    Data2 = GData[['intercept',i,'Y']]
    Data2[i] = list(map(float,Data2[i]))
    Data2['Y'] = list(map(int,Data2['Y']))
    #Data2.head()

    train_cols = Data2.columns[0:2]
    logit = sm.Logit(Data2['Y'], Data2[train_cols])
    # fit the model
    result = logit.fit()
    print(result.summary())
```

```python
x = ['intercept','人口數','性別比','扶養比','Y']
Data3 = GData[x]
#轉float
Data3['人口數'] = list(map(float,Data3['人口數']))
Data3['性別比'] = list(map(float,Data3['性別比']))
Data3['扶養比'] = list(map(float,Data3['扶養比']))
Data3['Y'] = list(map(int,Data3['Y']))

train_cols = Data3.columns[0:4]
logit = sm.Logit(Data3['Y'], Data3[train_cols])
# fit the model
result = logit.fit()
print(result.summary())
```
