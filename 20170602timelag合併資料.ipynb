```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt

#設定字體為標楷體'DFKai-sb':
from pylab import *
mpl.rcParams['font.sans-serif'] = ['DFKai-sb']
```

```python
f1 = open('H:\\LabData\\DATA\\蚊子資料(需要更新)\\2013-2015全區統整(時間編號).csv','r')
data_list = []
for i in f1.readlines():
    data = i.strip()
    data = data.split(',')
    data_list.append(data)
f1.close()
MosData = pd.DataFrame(data_list[1:],columns=data_list[0])
MosData.head()
```

```python
f1 = open('H:\\LabData\\DATA\\登革熱個案\\2013-2014登革熱發生數(時間編號).csv','r')
data_list = []
for i in f1.readlines():
    data = i.strip()
    data = data.split(',')
    data_list.append(data)
f1.close()
incidata = pd.DataFrame(data_list[1:],columns=data_list[0])
print(incidata.head())
Timelag = pd.DataFrame(list(map(int,list(incidata.ix[:,1]))))-12
Timelag = pd.DataFrame(list(map(str,list(Timelag.ix[:,0]))))
incidata['時間編號'] = Timelag
incidata.head()
```

```python
MergeData = pd.merge(MosData,incidata, on=['時間編號','區','里'])
#MergeData.head(10)
```
